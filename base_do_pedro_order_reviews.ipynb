```python
import pandas as pd
import numpy as np

# Rodrigo deu essa ideia de boas práticas na limpesa dele e só aproveitei
PATH = 'seeders/olist_order_reviews_dataset.csv'
CLEAN_PATH = 'clean_datasets/olist_order_reviews.csv'

# Deixei como df para padronizar os arquivos
df = pd.read_csv(PATH)


# display(df)
df.dtypes
df.isnull().any()
df.replace({True: 0, False: 1}, inplace=True)
df.isnull().any()
estatisticas = df.describe()
# print(estatisticas)

# Carregue o arquivo CSV
# df = pd.read_csv('olist_order_reviews_dataset.csv')

# Salve o DataFrame atualizado em um novo arquivo CSV
# df.to_csv('olist_order_reviews_dataset_preenchido.csv', index=False)
# Carregue o arquivo CSV
# df = pd.read_csv('olist_order_reviews_dataset_preenchido.csv')

# Verifique os tipos de dados atuais das colunas
# print("Tipos de dados antes da conversão:")
# print(df.dtypes)

# Converta as colunas para os tipos de dados corretos
#  Converter coluna 'review_creation_date' para datetime
df['review_creation_date'] = pd.to_datetime(df['review_creation_date'])
df['review_answer_timestamp'] = pd.to_datetime(df['review_answer_timestamp'])

# Salve o DataFrame atualizado em um novo arquivo CSV
# df.to_csv('olist_order_reviews_dataset_corrigido.csv', index=False)

# Verifique os tipos de dados após a conversão
# print("\nTipos de dados após a conversão:")
# print(df.dtypes)

# Carregue o arquivo CSV
# df = pd.read_csv('olist_order_reviews_dataset_corrigido.csv')

# Verifique se há linhas duplicadas
# print(f"Antes da remoção: {len(df)} linhas")

# Remova as duplicatas com base em todas as colunas
df.drop_duplicates(inplace=True)

# Verifique o número de linhas após a remoção
# print(f"Após a remoção: {len(df)} linhas")

# Salve o DataFrame atualizado em um novo arquivo CSV
# df.to_csv('olist_order_reviews_dataset_sem_duplicatas.csv', index=False)

# Carregue o arquivo CSV
# df = pd.read_csv('olist_order_reviews_dataset_sem_duplicatas.csv')

# Verifique os valores únicos em cada coluna para identificar possíveis erros de formatação
for coluna in df.columns:
    print(f"Valores únicos na coluna '{coluna}':")
    print(df[coluna].unique())

# Remova espaços em branco extras nas colunas de texto
colunas_texto = ['review_id', 'order_id', 'review_comment_title', 'review_comment_message']
for coluna in colunas_texto:
    df[coluna] = df[coluna].str.strip()

# Salve o DataFrame atualizado em um novo arquivo CSV
# df.to_csv('olist_order_reviews_dataset_corrigido.csv', index=False)

# Carregue o arquivo CSV com a codificação UTF-8
# df = pd.read_csv('olist_order_reviews_dataset_corrigido.csv', encoding='utf-8')


# Verifique os primeiros registros para garantir que a leitura foi bem-sucedida
# print(df.head())

# Salve o DataFrame atualizado em um novo arquivo CSV
# df.to_csv('olist_order_reviews_dataset_corrigido.csv', index=False, encoding='utf-8')


# Carregue o arquivo CSV
# df = pd.read_csv('olist_order_reviews_dataset_corrigido.csv')

 #Função para calcular o IQR
def calcular_iqr(serie):
    Q1 = serie.quantile(0.25)
    Q3 = serie.quantile(0.75)
    IQR = Q3 - Q1
    return IQR


# Calcular o IQR para cada coluna numérica
iqr = df.select_dtypes(include=['int64', 'float64']).apply(calcular_iqr)


# Identificar outliers
outliers = pd.DataFrame(np.zeros(df.shape, dtype=bool), columns=df.columns) #o dtype foi booleano, V ou F, mas pode ter outra visão
for col in iqr.index:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers[col] = (df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))

# display(outliers)

# Cria um novo arquivo csv já limpo e salva nesse caminho
df.to_csv(CLEAN_PATH, index=False)
```
